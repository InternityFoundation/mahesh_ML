# SMOTE (Synthetic Minority Over-sampling Technique
SMOTE is a technique which is used for oversampling of the dataset .That means if we have imbalance dataset or column like 
here we have in our target variable (some are having high counts and some are having low counts) so in oversampling we increase the minority values such that it makes the dataset balanced.

the way SMOTE works is -
# 1) first it find the nearest neighbour fo the vector points 
# 2) then it find the distance between the two nearest vector
# 3) then it multiply with the value between 0 and 1 
# 4) After multiplying it create a new synthetic vector point between those two vectors
# 5) And it keeps repeating the process and thats how the minority one's increases. 


```python
#oversampling using SMOTE 
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
print(x_train.shape,y_train.shape)
```

```python
model = SMOTE('minority')
x_train_res, y_train_res = model.fit_sample(x_train, y_train.ravel())
```
